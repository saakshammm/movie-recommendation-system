```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
movies = pd.read_csv("data/tmdb_5000_movies.csv")
credits = pd.read_csv("data/tmdb_5000_credits.csv")
```

```python
movies.head(2)
```

```python
credits.head(2)
```

```python
print(f"Shape of Movies: {movies.shape}")
print(f"Shape of Credits: {credits.shape}")
```

```python
movies = movies.merge(credits, on="title")
```

```python
movies.shape
```

```python
movies.columns
```

```python
movies = movies[["movie_id", "title", "overview", "genres", "keywords", "cast", "crew"]]
```

```python
movies.head(2)
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0]["genres"]
```

```python
import ast
def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i["name"])

    return l
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies.head(2)
```

```python
movies.iloc[0]["keywords"]
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies.head(2)
```

```python
movies.iloc[0]["cast"]
```

```python
def convert_cast(text):
    l = []
    count = 0
    for i in ast.literal_eval(text):
        if count > 3:
            l.append(i["name"])
        count +=1
    return l
```

```python
movies["cast"] = movies["cast"].apply(convert_cast)
movies.head(2)
```

```python
movies.iloc[0]["crew"]
```

```python
def fetch_director(text):
    l = []
    for i in ast.literal_eval(text):
        if i["job"] == "Director":
            l.append(i["name"])
            break

    return l
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
movies.head(2)
```

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
def remove_space(word):
    l = []
    for i in word:
        l.append(i.replace(" ", ""))
    return l
```

```python
movies["genres"] = movies["genres"].apply(remove_space)
movies["keywords"] = movies["keywords"].apply(remove_space)
movies["cast"] = movies["cast"].apply(remove_space)
movies["crew"] = movies["crew"].apply(remove_space)
```

```python
movies.head(2)
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head(2)
```

```python
movies.iloc[0]["tags"]
```

```python
df = movies[["movie_id", "title", "tags"]]
```

```python
df.head(2)
```

```python
df["tags"] = df["tags"].apply(lambda x: " ".join(x))
```

```python
df.head(2)
```

```python
df.iloc[0]["tags"]
```

```python
df["tags"] = df["tags"].apply(lambda x:x.lower())
```

```python
df.iloc[0]["tags"]
```

```python
df.shape
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
  l = []
  for i in text.split():
    l.append(ps.stem(i))

  return " ".join(l)
```

```python
df["tags"] = df["tags"].apply(stems)
```

```python
df.iloc[0]["tags"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = "english")
```

```python
vector = cv.fit_transform(df["tags"]).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarly= cosine_similarity(vector)
```

```python
similarly
```

```python
similarly.shape
```

```python
def recommend(movie):
  index = df[df["title"] == movie].index[0]
  distances = sorted(list(enumerate(similarly[index])), reverse = True, key = lambda x: x[1])
  for i in distances[1:6]:
    print(df.iloc[i[0]].title)
```

```python
recommend("Avatar")
```

```python
recommend("Spider-Man")
```
